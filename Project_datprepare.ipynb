```python
#ผู้ที่มาใช้สิทธิเลือกตั้งแบ่งเป็นภาค
#merge_data_constituencies_status with info_province_re
# Merge result_constituencies_PartyList with info_party_overview on 'party_id'
merged_df_1 = result_constituencies_status[['prov_id', 'turn_out']].merge(
    info_province_re[['prov_id', 'province','region_name']], left_on='prov_id', right_on='prov_id', how='inner'
)
ag = merged_df_1.groupby('region_name').sum()
ag_1 = ag.sort_values(by='turn_out',ascending=False)
ag_1.head(6)
```

```python
#ร้อยละของผู้ที่มาใช้สิทธิเลือกตั้งแบ่งเป็นภาค
#merge_data_constituencies_status with info_province_re
# Merge result_constituencies_PartyList with info_party_overview on 'party_id'
merged_df_1_1 = result_constituencies_status[['prov_id', 'turn_out']].merge(
    info_province_re[['prov_id', 'province','region_name']], left_on='prov_id', right_on='prov_id', how='inner'
)
ag1 = merged_df_1_1.groupby('region_name').mean()
ag1.head(6)
```

```python
#พรรคที่มีคะแนนโหวตมากที่สุดทั่วประเทศ
import pandas as pd

# Merge result_constituencies_PartyList with info_party_overview on 'party_id'
merged_df = result_constituencies_PartyList[['party_id', 'party_list_vote']].merge(
    info_party_overview[['id', 'name']], left_on='party_id', right_on='id', how='inner'
)

# Select only the desired columns
final_merged_df = merged_df[['party_id', 'party_list_vote', 'name']]

# Group by 'name' and calculate the sum of 'party_list_vote' for each group
party_vote_totals = final_merged_df.groupby('name')['party_list_vote'].sum().reset_index()

# Rename the 'party_list_vote' column to 'total_party_list_vote' if needed
party_vote_totals.rename(columns={'party_list_vote': 'total_party_list_vote'}, inplace=True)

party_vote_totals.sort_values(by='total_party_list_vote',ascending=False)


#party_vote_totals.to_csv('partylist_vote_totals.csv', index=False)
```

#แต่ละจังหวัดในภาคกลางเลือกพรรคอะไร

```python
find_ma = result_constituencies_Candidate.sort_values(by='mp_app_vote',ascending=False)
find_ma1 = find_ma.merge(info_constituency,how='left', on='cons_id')
find_ma2 = find_ma1.merge(info_province_re,how='left', on='prov_id')
find_ma3 = find_ma2.merge(region,how='left', on='region_name')
find_ma4 = find_ma3.merge(info_party_overview,how='left', left_on='party_id',right_on='id')
find_ma5 = find_ma4.merge(Candidate_constituency,how='left', on='mp_app_id')
```

```python
find_mashow =find_ma5[['province','name','mp_app_name','region_name','color']]
find_mashow['mp_app_vote_'] = find_ma5[['mp_app_vote']]
find_mashow
```

```python
#ผู้ที่มีคะแนนโหวตสูงสุดในภาคกลาง10อันดับ

use_V0 = find_mashow[find_mashow['region_name']== 'ภาคกลาง']
use_V77 =use_V0.drop_duplicates(subset=['mp_app_name'])
use_V78 =use_V0.drop_duplicates(subset=['province'])
use_V78.head(10)
```

```python
use_V77.head(10)
```

```python
#5พรรคอันดับแรกที่มีคะแนนสูงสุดในภาคกลาง
name_vote_totals = use_V77.groupby('name')['mp_app_vote_'].sum().reset_index()
name_vote_totals.sort_values(by='mp_app_vote_',ascending=False)

point5 = name_vote_totals.sort_values(by='mp_app_vote_',ascending=False)
point5

#point5.to_csv('name_vote_totals.csv', index=False)
```
